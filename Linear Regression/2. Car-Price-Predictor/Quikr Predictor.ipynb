# Makeing Model

```python
x = car.drop(columns='Price')
y = car['Price']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(x[['name','company','fuel_type']])
```

```python
ohe.categories_
```

```python
colum_trans = make_column_transformer((OneHotEncoder(categories= ohe.categories_),['name','company','fuel_type']), remainder= 'passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(colum_trans, lr)
```

```python
pipe.fit(x_train, y_train)
```

```python
pipe.predict(x_test)
```

```python
y_pred = pipe.predict(x_test)
```

```python
r2_score(y_test, y_pred)
```

```python
scores = []
for i in range(1000):
    x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.2, random_state= i)
    lr = LinearRegression()
    pipe = make_pipeline(colum_trans, lr)
    pipe.fit(x_train, y_train)
    y_pred = pipe.predict(x_test)
    scores.append(r2_score(y_test, y_pred))
```

```python
scores
```

```python
import numpy as np
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.2, random_state= np.argmax(scores))
lr = LinearRegression()
pipe = make_pipeline(colum_trans, lr)
pipe.fit(x_train, y_train)
y_pred = pipe.predict(x_test)
r2_score(y_test, y_pred)
```

```python
import pickle
pickle.dump(pipe,open('LinearRegressionModel.pk1','wb'))
```
